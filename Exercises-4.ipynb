### Define a year as a "Superman year" whose films feature more Superman characters than Batman. How many years in film history have been Superman years?

```python
cast_super = cast.groupby(["character", 'year']).size().loc["Superman"]
cast_super 
```

```python
cast_super_1 = cast_super.rename("Number of Superman characters")
cast_super_1
```

```python
cast_bat = cast.groupby(["character", 'year']).size().loc["Batman"]
cast_bat 
```

```python
cast_bat_1 = cast_bat.rename("Number of Batman characters")
cast_bat_1
```

```python
df = pd.merge(cast_super_1, cast_bat_1, on = 'year', how = 'outer')
df.fillna(0, inplace = True)
df
```

```python
df_1 = df.copy()
df_1["Superman Year"] = (df_1["Number of Superman characters"] > df_1["Number of Batman characters"]).astype(int)
df_1
```

```python
df_1["Superman Year"].sum()
```

### How many years have been "Batman years", with more Batman characters than Superman characters?

```python
df_2 = df.copy()
df_2["Batman Year"] = (df_2["Number of Batman characters"] > df_2["Number of Superman characters"]).astype(int)
df_2
```

```python
df_2["Batman Year"].sum()
```

### Count the number of actor roles each year and the number of actress roles each year over the history of film.

```python
cast["year"].min(), cast["year"].max()
```

```python
cast.groupby(['year', 'type']).size()
```

```python
df_3 = cast.groupby(['year', 'type']).size().unstack()
df_3
```

```python
df_3.reindex(range(1894, 2027)).fillna(0)
```

### Find the difference between the number of actor roles each year and the number of actress roles each year over the history of film.

```python
df_4 = df_3.reindex(range(1894, 2027)).fillna(0)
df_4["difference"] = df_4["actor"] - df_4["actress"]
df_4["difference"]
```

```python
df_4.loc[2000]
```

### What is the fraction of roles that have been 'actor' roles each year in the history of film.

```python
df_4
```

```python
df_4["fraction of actor roles"] = (df_4["actor"] / (df_4["actor"] + df_4["actress"]))
df_4.fillna(0, inplace = True)
df_4
```

### What is the fraction of supporting (n=2) roles that have been 'actor' roles each year in the history of film.

```python
cast.groupby(['year', 'type', 'n']).size().loc[:, :, 2]
```

```python
df_5 = cast.groupby(['year', 'type', 'n']).size().loc[:, :, 2].unstack()
df_5
```

```python
df_6 = df_5.reindex(range(1894, 2027)).fillna(0)
df_6
```

```python
df_6["fraction of actor roles"] = (df_6["actor"] / (df_6["actor"] + df_6["actress"]))
df_6.fillna(0, inplace = True)
df_6
```
